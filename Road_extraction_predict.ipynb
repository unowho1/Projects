```python
from keras.models import load_model
import cv2
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import os

def pred1(img_name, model, shape=128):
    img = cv2.imread(img_name)
    img = cv2.cvtColor(img, cv2.COLOR_RGBA2RGB)
    img = cv2.resize(img, (shape, shape))
    predictions = model.predict(np.array([img]))
    return predictions[0], img

def Plotter(img, predMask):
    plt.figure(figsize=(9,9))
    
    plt.subplot(1,3,1)
    plt.imshow(img)
    plt.title('Aerial image')
    
    plt.subplot(1,3,2)
    plt.imshow(predMask)
    plt.title('Predicted Routes')

dir_path = "C:/Users/Sujal/Documents/Submissions/PSC/Innovative Assignment/Satellite images"
model_name = load_model("Model.h5")

output_dir = "Predicted routes"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

for filename in os.listdir(dir_path):
    if filename.endswith(".png"):
        img_path = os.path.join(dir_path, filename)
        sp, a = pred1(img_path, model_name)
        Plotter(a, sp)
        output_path = os.path.join(output_dir, "prediction_" + filename)
        cv2.imwrite(output_path, sp*255)
```
